### Subpoint b:

```python
import numpy as np
import matplotlib.pyplot as plt

a0 = 1.0
r = np.linspace(0, 2, 1000)

def psi_1s(r, Z):
    return (1 / np.sqrt(np.pi * a0**3)) * Z**1.5 * np.exp(-Z * r / a0) # siehe folien

plt.plot(r, psi_1s(r, 3), label="Li")
plt.plot(r, psi_1s(r, 11), label="Na")
plt.plot(r, psi_1s(r, 19), label="K")

plt.xlabel("r")
plt.ylabel(r"$\psi_{1s}(r)$")
plt.legend()
plt.show()
```
